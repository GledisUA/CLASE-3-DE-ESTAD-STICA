```python
# prompt: crea una tabla de frecuencias y el histograma con su poligono de frecuencia con la variable PESO

# Paso 10. Tabla de frecuencias para la variable PESO
freq_table_peso = df['PESO'].value_counts().reset_index()
freq_table_peso.columns = ['PESO', 'Frecuencia']
freq_table_peso['Frecuencia Relativa'] = freq_table_peso['Frecuencia'] / freq_table_peso['Frecuencia'].sum()
freq_table_peso['Frecuencia Acumulada'] = freq_table_peso['Frecuencia'].cumsum()
freq_table_peso['Frecuencia Relativa Acumulada'] = freq_table_peso['Frecuencia Relativa'].cumsum()

print("Tabla de Frecuencias para PESO:")
print(freq_table_peso)

# Paso 11. Histograma y polígono de frecuencia para la variable PESO
plt.figure(figsize=(10, 6))
# Histograma
sns.histplot(df['PESO'], kde=False, bins=20, color='skyblue', label='Histograma') # Adjust bins as needed
plt.title('Histograma de la variable PESO')
plt.xlabel('PESO')
plt.ylabel('Frecuencia')

# Polígono de frecuencia
# Calculate bin centers for the polygon
hist, bin_edges = np.histogram(df['PESO'], bins=20) # Use the same bins as the histogram
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Add points for the polygon at the start and end with frequency 0
bin_centers = np.insert(bin_centers, 0, bin_edges[0] - (bin_centers[0] - bin_edges[0]))
bin_centers = np.append(bin_centers, bin_edges[-1] + (bin_edges[-1] - bin_centers[-1]))
hist = np.insert(hist, 0, 0)
hist = np.append(hist, 0)

plt.plot(bin_centers, hist, marker='o', color='red', linestyle='-', label='Polígono de Frecuencia')

plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```
